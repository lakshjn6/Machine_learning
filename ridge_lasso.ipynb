```python
import numpy as np
```

```python
from sklearn.linear_model import Lasso
```

```python
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\BIT\\OneDrive\\Desktop\\placement.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
```

```python
x.head()
```

```python
y=df.iloc[:,-1]
```

```python
y.head()
```

```python
y=y.values.reshape(-1,1)
```

```python
y
```

```python
x=x.values.reshape(-1,1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2)
```

```python
lasso_reg=Lasso()
```

```python
parameter={'alpha':[0.2,0.4,0.6,2,3,4,5,6,9,10,14,19,20,34]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lasso_cv=GridSearchCV(lasso_reg,parameter,scoring='neg_mean_squared_error',cv=5)
```

```python
lasso_cv.fit(x_train,y_train)
```

```python
print(lasso_cv.best_params_)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,lasso_cv.predict(x_test))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_train,y_train)
plt.plot(
    x_test,lasso_cv.predict(x_test),color='r'
)

```
