```python
from sklearn.linear_model import LogisticRegression
```

```python
classification=LogisticRegression()
```

```python
df['person_education']=df['person_education'].map({'Master':0,'High School':1, 'Bachelor':2, 'Associate':3, 'Doctorate':4})
```

```python
df['person_home_ownership'].unique()
```

```python
df['person_home_ownership']=df['person_home_ownership'].map({'RENT':1, 'OWN':2, 'MORTGAGE':3, 'OTHER':4})
```

```python
df['loan_intent'].unique()
```

```python
df['loan_intent']=df['loan_intent'].map({'PERSONAL':1, 'EDUCATION':2, 'MEDICAL':3, 'VENTURE':4, 'HOMEIMPROVEMENT':5,
       'DEBTCONSOLIDATION':5})
```

```python
df.head()
```

```python
df['previous_loan_defaults_on_file']=df['previous_loan_defaults_on_file'].map({'No':0,'Yes':1})
```

```python
x=df[['person_age','person_gender','person_education','person_income','person_emp_exp','person_home_ownership','loan_amnt','loan_intent','loan_int_rate','loan_percent_income','cb_person_cred_hist_length','credit_score','previous_loan_defaults_on_file']]
```

```python
x.head()
```

```python
x=x.values.reshape(-1,13)
```

```python
x
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2)
```

```python
classification.fit(x,y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,classification.predict(x_test))
```

```python
parameter={'penalty':['l1', 'l2', 'elasticnet'],'C':[1,2,3,4,5,6,10,15,20],'max_iter':[100]}
```

```python
classification_regression=GridSearchCV(classification,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classification_regression.fit(x_train,y_train)
```

```python
print(classification_regression.best_params_)
```

```python
accuracy_score(y_test,classification_regression.predict(x_test))
```
